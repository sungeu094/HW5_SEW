```python
import csv

def main():
    f = open('q4.csv')
    data = csv.reader(f)
    header = next(data)

    max_passengers = [-99999, -99999, -99999, -99999]
    max_station = ['', '', '', '']
    min_passengers = [999999, 999999, 999999, 999999]
    min_station = ['', '', '', '']

    for row in data:
        for i in range(4, 6):
            row[i] = int(row[i].replace(',', ''))  # 쉼표 제거하여 정수로 변환

        total_passengers = row[4] + row[5]
        line = row[1]

        if line == '1호선':
            if max_passengers[0] < total_passengers:
                max_passengers[0] = total_passengers
                max_station[0] = row[3]
            if min_passengers[0] > total_passengers:
                min_passengers[0] = total_passengers
                min_station[0] = row[3]
        elif line == '2호선':
            if max_passengers[1] < total_passengers:
                max_passengers[1] = total_passengers
                max_station[1] = row[3]
            if min_passengers[1] > total_passengers:
                min_passengers[1] = total_passengers
                min_station[1] = row[3]
        elif line == '3호선':
            if max_passengers[2] < total_passengers:
                max_passengers[2] = total_passengers
                max_station[2] = row[3]
            if min_passengers[2] > total_passengers:
                min_passengers[2] = total_passengers
                min_station[2] = row[3]
        elif line == '4호선':
            if max_passengers[3] < total_passengers:
                max_passengers[3] = total_passengers
                max_station[3] = row[3]
            if min_passengers[3] > total_passengers:
                min_passengers[3] = total_passengers
                min_station[3] = row[3]

    print("*** Subway Report for Seoul on March 2023 ***")
    print("Line 1")
    print("Busiest Station:", max_station[0], "(", max_passengers[0], ")")
    print("Least used Station:", min_station[0], "(", min_passengers[0], ")")
    print("Line 2")
    print("Busiest Station:", max_station[1], "(", max_passengers[1], ")")
    print("Least used Station:", min_station[1], "(", min_passengers[1], ")")
    print("Line 3")
    print("Busiest Station:", max_station[2], "(", max_passengers[2], ")")
    print("Least used Station:", min_station[2], "(", min_passengers[2], ")")
    print("Line 4")
    print("Busiest Station:", max_station[3], "(", max_passengers[3], ")")
    print("Least used Station:", min_station[3], "(", min_passengers[3], ")")

    f.close()


if __name__ == '__main__':
    main()
```
